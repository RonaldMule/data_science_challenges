### 10. Cumulative sum and Rolling Calulations

```python
apple_change = np.log(apple.Close /apple.Close.shift(1))
google_change = np.log(google.Close / google.Close.shift(1))
```

```python
apple_change.cumsum().plot() 
google_change.cumsum().plot()
```

### 11. create a new dataFrame and save to csv

```python
new = pd.DataFrame({'AAPL': apple_change[1:], 'GOOG': google_change[1:]})
new.to_csv('stockug.csv')
```
